```python
#using dictionaries
#Bar plot
dict={"peanut butter":10,"Cashew butter":12,"Cheese":8}
fig,ax=plt.subplots()
ax.bar(dict.keys(),dict.values());
ax.set(title="Sales",xlabel="Things",ylabel="Cost");
```

```python
#line plot
fig,ax=plt.subplots()
ax.plot(dict.keys(),dict.values());
ax.set(title="Sales",xlabel="Things",ylabel="Cost");
```

```python
#scatter plot
fig,ax=plt.subplots()
ax.scatter(dict.keys(),dict.values());
ax.set(title="Sales",xlabel="Things",ylabel="Cost");
```

```python
#horizontal bar plot
fig,ax=plt.subplots()
ax.barh(list(dict.keys()),list(dict.values()));
ax.set(title="Sales",xlabel="Things",ylabel="Cost");
```

```python
#subplot method 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(a,a/2);
ax2.scatter(a,np.exp(a));
ax3.hist(n);
ax4.bar(x,y);
ax1.set(title="line",xlabel="values",ylabel="terms");
ax2.set(title="scatter",xlabel="values",ylabel="terms");
ax3.set(title="histogram",xlabel="values",ylabel="terms");
ax4.set(title="bar",xlabel="values",ylabel="terms");
```

```python
#subplot method 2
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(a,a/2);
ax[0,1].scatter(a,np.exp(a));
ax[1,0].hist(n);
ax[1,1].bar(x,y);
ax[0,0].set(title="line",xlabel="values",ylabel="terms");
ax[0,1].set(title="scatter",xlabel="values",ylabel="terms");
ax[1,0].set(title="histogram",xlabel="values",ylabel="terms");
ax[1,1].set(title="bar",xlabel="values",ylabel="terms");
```

```python
data=pd.read_csv("heart-disease.csv")
data.head()
```

```python
data["age"].plot.hist(bins=10);
```

```python
data.plot(figsize=(10,30),subplots=True);
```

```python
# OO method with pyplot method
fig,ax=plt.subplots(figsize=(10,6))
over_50=data[data["age"]>50]
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax);
#ax.set_xlim([45,100]);
```

```python
# OO method from scratch
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])

#customize
ax.set(title="Heart Disease Levels",xlabel="Age",ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(),title="Target");

#hr line
ax.axhline(over_50["chol"].mean());
```

```python
over_50.head()
```

```python
#also customizing the plots

fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10))
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter");
ax0.set(title="heart dis",xlabel="age",ylabel="chol");
ax0.legend(*scatter.legend_elements(),title="Legend");
ax0.set_xlim([50,80]);
scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"],cmap="winter");
ax1.set(title="heart dis",xlabel="age",ylabel="thalach");
ax1.legend(*scatter.legend_elements(),title="Legend");
ax1.set_xlim([50,80]);
ax1.set_ylim([60,200]);
fig.suptitle("Heart diseases analysis",fontsize=16,fontweight="bold");
```
